```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline

file = 'stats2.tsv'

# Initialize dictionaries to store data for pie chart
share_data = {
    'fsb': [],
    'typo': [],
    'emo': []
}

df = pd.read_csv(f'./{file}', sep='\t')

def draw_cap(ax, y_column, name):
    x_column = 'cover size'

    df_sorted = df.sort_values(x_column)

    x = df_sorted[x_column]
    y = df_sorted[y_column]

    mean_bits = np.mean(df_sorted[y_column.replace('capacity', 'bits')])
    std_bits = np.std(df_sorted[y_column.replace('capacity', 'bits')])
    mean_x = np.mean(x)
    std_x = np.std(x)
    mean_y = np.mean(y)
    std_y = np.std(y)

    ax.set_title(name)
    ax.scatter(x, y, label='Samples')
    ax.grid(True)
    ax.set_xlabel(x_column)
    ax.set_ylabel('Capacity')

    ax.set_ylim(-2, 50)
    ax.legend()

    print(f'Mean Bits: {mean_bits:.2f}\nStd Bits: {std_bits:.2f}')
    print(f'Mean {x_column}: {mean_x:.2f}\nStd {x_column}: {std_x:.2f}')
    print(f'Mean Ratio: {mean_y:.2f}\nStd Ratio: {std_y:.2f}')

# Create a 1x4 grid of plots
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

# Call the plotting functions for each subplot
draw_cap(axes[3], 'capacity', 'Stegasus')
draw_cap(axes[2], 'typo capacity', 'Typoceros')
draw_cap(axes[1], 'emo capacity', 'Emojier')
draw_cap(axes[0], 'fsb capacity', 'Frustratingly Simple Bert')

# Adjust layout to pack the plots tightly
plt.tight_layout()

# Display the plots
plt.show()
```
